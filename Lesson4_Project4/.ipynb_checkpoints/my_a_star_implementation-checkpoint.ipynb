```python
import math


def shortest_path_function(mymap, start, goal):
    # Initialize data containers
    frontier, explored_list, path = [], [], []
    track_parent = {}

    # Add start node and its cost to frontier dictionary
    frontier.append([0, start])
    track_parent[start] = None

    # Start exploring the map
    while len(frontier) > 0:
        frontier.sort()
        frontier.reverse()
        current_node = frontier.pop()[1]
        explored_list.append(current_node)

        neighbor = mymap.roads[current_node]
        for i in neighbor:
            exists_flag = False
            cost_g = math.sqrt((mymap.intersections[current_node][0] - mymap.intersections[i][0])**2 + (mymap.intersections[current_node][1] - mymap.intersections[i][1])**2)
            cost_h = math.sqrt((mymap.intersections[goal][0] - mymap.intersections[i][0])**2 + (mymap.intersections[goal][1] - mymap.intersections[i][1])**2)
            cost_f = cost_g + cost_h

            if i in explored_list:
                continue

            for indx, ele in enumerate(frontier):
                if i == ele[1]:
                    exists_flag = True
                    if ele[0] > cost_f:                # found a better path to reach node i
                        frontier[indx][0] = cost_f     # update with lower cost
                        track_parent[i] = current_node # override bad parent
                    break
            if not exists_flag:
                frontier.append([cost_f, i])
                track_parent[i] = current_node

    # Get the shortest path out
    path.append(goal)
    parent = track_parent[goal]
    while (not parent is None):
        path.append(parent)
        parent = track_parent[parent]
    
    # reserse the path order to start => goal
    path.reverse()
    return path
    # print(track_parent)
    # print(track_parent[34])
    # print(track_parent[12])
    # print(track_parent[37])
    # print(track_parent[16])
    # print(track_parent[5])

start, goal, mymap = 5, 34, map_40
shortest_path = shortest_path_function(mymap, start, goal)
print(shortest_path)
```
